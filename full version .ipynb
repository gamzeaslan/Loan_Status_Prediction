```python
import pandas as pd
```

```python
data=pd.read_csv("credit_test.csv")
data.head()
```

```python
#gereksiz kolonların atılması
data=data.drop(["Loan ID","Customer ID"],axis=1)
data=data.reset_index(drop=True)
data.head()
```

```python
#eksik değerlerin yüzdeliklerin belirlenmesi
data.isnull().sum()/len(data["Purpose"])*100
```

```python
#Months since last delinquent  değişkenindeki eksik değer yüzdesi
#%54 olduğundan  bu kolonu siliyorum
data=data.drop("Months since last delinquent",axis=1)
data.head()
```

```python
#eksik değerlerin rastgeliliğin hesaplanması:
import missingno as msno
msno.matrix(data)
```

```python
#Credit Score ile Annual Income arasında bir ilişki var galiba çünkü 
#eksik değerler ortak eksik değerler var 
```

```python
#kategorik değişkenlerdeki eksik değerlerin silinmesi
data.info()
```

```python
data.isnull().sum()
```

```python
#kategorik değişkenlerdeki null değerlerin silinmesi
data.dropna(subset=['Term', 'Years in current job',
                    "Home Ownership","Purpose"], inplace=True)
data=data.reset_index(drop=True)
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
#diğer eksik değerlerin KNN kullanılarak doldurulması
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=5)
numeric_columns=imputer.fit_transform(data[["Credit Score","Annual Income",
                                           "Bankruptcies","Tax Liens"]])
numeric_df=pd.DataFrame(data=numeric_columns,
                       columns=["Credit Score","Annual Income","Bankruptcies","Tax Liens"])
numeric_df.head()
```

```python
numeric_df.isnull().sum()
```

```python
numeric_df.shape
```

```python
numeric_df.tail()
```

```python
data.drop(["Credit Score","Annual Income","Bankruptcies","Tax Liens"],axis=1,inplace=True)
data.head()
```

```python
data=pd.concat([data,numeric_df],axis=1)
data.isnull().sum()
```

```python
#Years in current job kolonun düzeltilmesi:
def year_job(data):
    term=data.replace("years","").replace("+","").replace("year","").replace("<","").strip()
    return term
data["Years in current job"]=data["Years in current job"].apply(year_job)
data.head()
```

```python
#Term kolonunun düzeltilmesi:
def Term(data):
    if(data.lower()=="short term"):
        data="0"
    elif(data.lower()=="long term"):
        data="1"
    return data
data["Term/Long Term"]=data["Term"].apply(Term)
data=data.drop("Term",axis=1)
data.head()
```

```python
dummies_home_ownership=pd.get_dummies(data["Home Ownership"])
dummies_purpose=pd.get_dummies(data["Purpose"])
```

```python
data=data.drop(["Home Ownership","Purpose"],axis=1)
data=pd.concat([data,dummies_home_ownership,dummies_purpose],axis=1)
data.head()
```

```python
data.isnull().sum()
```

```python
#değişkenlerin tiplerinin kontrol edilmesi
data.info()
```

```python
data["Years in current job"]=data["Years in current job"].astype(int)
data["Term/Long Term"]=data["Term/Long Term"].astype(int)
```

```python
data.info()
```

```python
data.to_csv("cleaning data.csv")
```

```python
#kmeans kümeleme algortimasını kullanarak kümele
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
labels=kmeans.labels_
centers=kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
x=data
y=labels
score = silhouette_score(x, kmeans.labels_)
print(score)
```
